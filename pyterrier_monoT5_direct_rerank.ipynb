```python
import pandas as pd
import pyterrier as pt
pt.init()
```

```python
from pyterrier_t5 import MonoT5ReRanker
monoT5 = MonoT5ReRanker()
```

```python
example_input = pd.DataFrame([
    ['q1', 'chemical reactions', 'd1', 'excitation mechanisms of the oxygen emission in the upper atmosphere\nphotochemical reactions affected by mass motions and direct excitation\ndue to mass motions are examined'
],
    ['q1', 'chemical reactions', 'd2', 'ion neutral reactions  a list is given of reactions which may control\nthe chemical composition of positive ions in the ionosphere  under\ncertain assumptions about reaction rate coefficients nitrogen is\nsubstantially dissociated in the f region to the theoretical conclusions\nof other investigators']
    ], columns=['qid', 'query', 'docno', 'text'])
example_input
```

```python
monoT5.transform(example_input)

# ranking is defined by the _descending_ score values, 
# or, equivalently the _ascending_ rank
# NOT by the ordering of the dataframe

# so, we see that d2 is estimated to be more relevant to the query
```
